```
#Ejemplo 3
#con dos iguales comparas valores y uno asignas valor
email = False
for i in"ariel@unam.mx":
  if (i=="@"):
    email=True
if email == True:
  print("El email es correcto")

else:
  print("El correo no es correcto")
```
